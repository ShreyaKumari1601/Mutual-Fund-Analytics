```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data_raw/mutual_funds_data.csv")
```

### Observation

- The `date` column was originally stored as a string (`object`).
- It has been successfully converted to the `datetime64[ns]` datatype.
- This enables time-based analysis such as yearly trends, monthly summaries, and date filtering.
