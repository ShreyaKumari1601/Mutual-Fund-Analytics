# Mutual Fund Analytics Dashboard

## Notebook 4 : Feature Engineering

### Objective

Create new business features from the cleaned mutual fund dataset to support SQL analysis and Power BI dashboard.

### Business Goal

Generate useful metrics such as year, month, daily return, rolling average, and volatility to improve fund performance analysis.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data_cleaned/cleaned_mutual_funds_data.csv")
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["Year"] = df["date"].dt.year
```

```python
df["Year"] = df["date"].dt.year
```

```python
df[["date","Year"]].head()
```

```python
df["Month"] = df["date"].dt.month_name()
```

```python
df[["date","Month"]].head()
```

```python
df["Quarter"] = df["date"].dt.quarter
```

```python
df["Day"] = df["date"].dt.day
```

```python
df["Day_Name"] = df["date"].dt.day_name()
```

```python
df = df.sort_values(["scheme_name","date"])
```

```python
df["Daily_Return"] = (
    df.groupby("scheme_name")["net_asset_value"]
      .pct_change()
)
```

```python
df[["scheme_name",
    "date",
    "net_asset_value",
    "Daily_Return"]].head(15)
```

```python
df["Rolling_NAV"] = (
    df.groupby("scheme_name")["net_asset_value"]
      .transform(lambda x: x.rolling(7).mean())
)
```

```python
df["Volatility"] = (
    df.groupby("scheme_name")["Daily_Return"]
      .transform(lambda x: x.rolling(30).std())
)
```

```python
df.to_csv(
    "../data_cleaned/mutual_fund_featured.csv",
    index=False
)

print("Feature engineered dataset saved successfully.")
```
